```python
import pandas as pd
from tabula.io import read_pdf

DATADICT='../notes/temp.pdf'
pd.options.display.max_columns = None
pd.set_option('display.max_colwidth', None)
```

# assignments
the assignments table describes a house sit, agreed upon by homeowner and housesitter.

```python
assignments_desc_df = (
    read_pdf(DATADICT, pages=[1])[0]
    .rename(columns={'Field':'field', 'Field Description':'description'})
)
# for the purposes of an entity relationship diagram, here are the columns containing foreign keys:
assignments_desc_df[assignments_desc_df['field'].str.endswith('_id')]
```

# profiles

the profiles table is spread across multiple pages which are all read in separately and need to be concatenated. 

also, the header is missing from all but the first page entry, meaning a data row is mistaken for the header.

```python
# read in the pages
profiles_desc_df_list = read_pdf(DATADICT, pages=[2,3,4])

top_row=list(profiles_desc_df_list[1].columns)
single_row=list(profiles_desc_df_list[2].columns)
profiles_desc_df=(
    pd.concat(
        [
            profiles_desc_df_list[0]
                .rename(columns={'Field':'field', 'Field Description':'description'}),
            pd.DataFrame({'field':[top_row[0]], 'description': [top_row[1]]}),
            profiles_desc_df_list[1]
                .rename(columns={top_row[0]:'field', top_row[1]:'description'}),
            pd.DataFrame({'field':[single_row[0]], 'description': [single_row[1]]})
        ]
    )
    .reset_index()
    .drop(columns=['index'])
)
# for the purposes of an entity relationship diagram, here are the columns containing foreign keys:
profiles_desc_df[profiles_desc_df['field'].str.endswith('_id')]
```

# listings 
this table is spread across two pages as well (p. 5 and p. 6), and the first row of the second page is a continuation of the last row of the first page.

```python
listing_desc_df_list = read_pdf(DATADICT, pages=[5,6])
top_row=list(listing_desc_df_list[1].columns)
listing_desc_df = (
    pd.concat(
        [
            listing_desc_df_list[0]
                .rename(columns={'Field':'field', 'Field Description':'description'}),
            pd.DataFrame({'field':[top_row[0]], 'description': [top_row[1]]}),
            listing_desc_df_list[1]
                .rename(columns={top_row[0]:'field', top_row[1]:'description'})
        ]
    )
    .reset_index()
    .drop(columns=['index'])
)

# for the purposes of an entity relationship diagram, here are the columns containing foreign keys:
listing_desc_df[listing_desc_df['field'].str.endswith('_id')]
```
